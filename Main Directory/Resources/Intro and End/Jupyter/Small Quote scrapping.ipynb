```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
pd.set_option('display.width', 200)
pd.set_option('display.max_columns', 10)

url = "https://www.codeofliving.com/blog/55-powerful-short-quotes-sayings-life"

source = requests.get(url).text
soup = BeautifulSoup(source, 'html.parser')

quote_all = soup.find_all("li", {'class': None})

all_text = [item.get_text() for item in quote_all if item.get_text()!=""]
all_text = [item.split("–") for item in all_text]

my_list = []
for i in all_text:    
    dict= {}
    try:
        dict["Text"]= i[0].strip()
        dict["Author"]= i[1].strip()
        
    except:
        pass
    my_list.append(dict)
    
my_list
```

```python
df = pd.DataFrame(my_list)
df
```
